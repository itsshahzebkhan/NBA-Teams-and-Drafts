```python
#EDA Questions= 1949 to 2020, top 5?
q1="""SELECT season,Count(Distinct team_id) as Count_teams
from draft_history
group by season
order by Count_teams desc limit 5"""
r1=pd.read_sql(q1,conn)
print(r1)
```

```python
#top 10 teams
q2="""SELECT team_name, Count(distinct person_id) as total_drafted
FROM draft_history
GROUP BY team_name
order by total_drafted
Limit 10"""
r2=pd.read_sql(q2,conn)
print(r2)
```
